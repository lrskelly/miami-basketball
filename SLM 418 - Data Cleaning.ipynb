# **Pre-Cluster Analysis Data Cleaning**

### **Import and Examine Raw Data**

```python
# Import pandas
import pandas as pd

# Read in csv's
defense = pd.read_csv('MAC overall D 2014-2025 - MAC overall D 2014-2025.csv')
general = pd.read_csv('MAC O 2014-2025 SRCBB - MAC O 2014-2025 SRCBB.csv')
```

```python
# Check structure of defense data
defense.info()
```

```python
# Add 'Opp ' before the name of each field from the 6th index to the last
defense.columns = list(defense.columns[:6]) + ['Opp ' + col for col in defense.columns[6:]]
defense.head()
```

```python
general.info()
```

```python
general.head()
```

### **Fix Naming Convention Inconsistencies**

```python
# Make records in 'Player' lowercase for both files
defense['Player'] = defense['Player'].str.lower()
general['Player'] = general['Player'].str.lower()
```

```python
# For 'Player' in defense, eliminate '.' in any record
defense['Player'] = defense['Player'].str.replace('.', '')
general['Player'] = general['Player'].str.replace('.', '')
```

```python
# Replace "  " with ' ' for records in 'Player'
defense['Player'] = defense['Player'].str.replace('  ', ' ')
general['Player'] = general['Player'].str.replace('  ', ' ')
```

```python
# Replace ' jr' with ''
defense['Player'] = defense['Player'].str.replace(' jr', '')
general['Player'] = general['Player'].str.replace(' jr', '')
```

```python
# Replace ' iii' with ''
defense['Player'] = defense['Player'].str.replace(' iii', '')
general['Player'] = general['Player'].str.replace(' iii', '')
```

```python
# Replace ' iv' with ''
defense['Player'] = defense['Player'].str.replace(' iv', '')
general['Player'] = general['Player'].str.replace(' iv', '')
```

```python
# Manually change records as necessary to match names across files
defense['Player'] = defense['Player'].str.replace('juanse gorosito', 'juan sebastian gorosito')
defense['Player'] = defense['Player'].str.replace('bryce bultman', 'bryson bultman')
defense['Player'] = defense['Player'].str.replace('jermal davis', 'jemal davis')
defense['Player'] = defense['Player'].str.replace('ray j dennis', 'rayj dennis')
defense['Player'] = defense['Player'].str.replace('keshawn william', 'keshawn williams')
defense['Player'] = defense['Player'].str.replace('zid powell', 'yazid powell')
defense['Player'] = defense['Player'].str.replace('ben vander-plas', 'ben vander plas')
defense['Player'] = defense['Player'].str.replace('chinedu okanu', 'chinedu kingsley okanu')
defense['Player'] = defense['Player'].str.replace("travon fagan", "tra'von fagan")
defense['Player'] = defense['Player'].str.replace("hunter darweshi", 'darweshi hunter')
defense['Player'] = defense['Player'].str.replace("loren cristian jackson", 'loren jackson')
defense['Player'] = defense['Player'].str.replace("trey simons", 'troy simons')
defense['Player'] = defense['Player'].str.replace("chanell banks", 'channel banks')
defense['Player'] = defense['Player'].str.replace("nathan scott", 'nate scott')
defense['Player'] = defense['Player'].str.replace("spencer litlteson", 'spencer littleson')
defense['Player'] = defense['Player'].str.replace("emmanuel olojakpoe", 'emmanuel olojakpoke')
defense['Player'] = defense['Player'].str.replace("willie magnum", 'willie mangum')
defense['Player'] = defense['Player'].str.replace("tavian dunn martin", 'tavian dunn-martin')
defense['Player'] = defense['Player'].str.replace("delvin dickerson", "del'vin dickerson")
defense['Player'] = defense['Player'].str.replace("lekan ajayi", 'olalekan ajayi')
defense['Player'] = defense['Player'].str.replace("joe reece", 'joseph reece')
```

### **Merge Files**

```python
# Join defense to general using the 'Player' and 'Year' fields
final = general.merge(defense, on=['Player', 'Season'], how='left')
final.head()
```

### **Filter Out Low-Value Records and Fields**

```python
# Eliminate all players with 'MP < 10'
final = final[final['MP'] >= 10]

# Eliminate all players with 'G' < 10
final = final[final['G'] >= 10]
```

```python
# Drop unuseful fields
final = final.drop(columns=['Rk', 'Awards', 'Player-additional', 'Rank', 'Jersey #', 'Team_y', 'GP'])
```

### **Impute Missing Values**

```python
# Examine data structure to check for missing values
final.info()
```

```python
# Check records in final where 'Pos' is missing
final[final['Pos'].isnull()]
```

```python
# For records where 'Player' is 'james thompson', fill 'Pos' with 'C' after performing research to find his correct position
final.loc[final['Player'] == 'james thompson', 'Pos'] = 'C'
```

```python
# Check records in final where '3P%' is missing
final[final['3P%'].isnull()].T
```

```python
# Check records in final where 'FT%' is missing
final[final['FT%'].isnull()].T
```

```python
# Check records in final where 'Opp 3 FG%' is missing
final[final['Opp 3 FG%'].isnull()].T
```

```python
# When '3P%' is missing or null, imputate with 0 since it appears that these records
# have this field empty because the player never attempted a three point field goal
final['3P%'] = final['3P%'].fillna(0)

# When 'FT%' is missing or null, imputate with 0 since the only record where FT%
# is missing is for a player that never shot a free throw
final['FT%'] = final['FT%'].fillna(0)

# When 'Opp 3 FG%' is missing or null, imputate with 0 since the only record where
# this field is missing is for a player who never defended a three point attempt
final['Opp 3 FG%'] = final['Opp 3 FG%'].fillna(0)
```

```python
# Check for missing values in final
final.isnull().sum().T
```

### **Final Checks**

```python
# Examine data structure
final.info()
```

```python
# Rename 'Team_x' to 'Team
final = final.rename(columns={'Team_x': 'Team'})
```

### **Export Cleaned Data**

```python
# Save final into a .csv file called 'master.csv'
final.to_csv('master.csv', index=False)
```
